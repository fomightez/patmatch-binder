## Patmatch 1.2 README

About
----

This directory contains the program used by PatMatch to do the actual pattern  
matching.  The program is written in C and Perl.  It has been tested on Solaris  
and Linux systems only.  The program can only be used from a terminal window.  

Note: PatMatch was also tested on MacOSX by Dr. Rodolfo Chaparro. The  
developer tools from the Apple website  
(http://developer.apple.com/tools) must be downloaded in order to  
compile Patmatch on MacOSX.  

PatMatch 1.x provides a more powerful pattern matching than PatMatch in that  
it allows patterns containing wildcards to support mismatch options, and it now  
supports patterns containing both regular expressions and mismatch options.  

This program uses a modified version of the NR-grep software for pattern  
matching by G. Navarro.  The original version of NR-grep can be found at  
http://www.dcc.uchile.cl/~gnavarro/software/  

This program was developed by Thomas Yan and Danny Yoo.  

For more information, please see the paper:  
Yan, T, Yoo, D, Berardini, T, Mueller LA, Weems, D, Weng, S, Cherry JM,  
Rhee, SY (2005) PatMatch: a program for finding patterns in peptide and  
nucleotide sequences.  Nucleic Acids Research. 33 Web Server issue:W262-6.  
Link:  
http://nar.oxfordjournals.org/cgi/reprint/33/suppl_2/W262?ijkey=ZEc5zYlbTzvtJH1&keytype=ref


Perl Modules Needed
-------------------

FindBin  
Parse::RecDescent


Installation
------------

In a terminal window, type:  
gzip -d patmatch_1.2.tar.gz  

To extract the decompressed tar file's contents, type:  
tar -xvf patmatch_1.2.tar  

This creates a new directory called patmatch_1.2.  cd into this directory and  
compile the program by typing:  
make


Usage
-----

Preparing FASTA datasets for PatMatch.  This must be done before running  
PatMatch:

patmatch.pl accepts unjustified FASTA files as sequence files.  An  
unjustified FASTA file is one where each sequence is in one line.  A FASTA  
file can be unjustified using the unjustify_fasta.pl program.  This is   
necessary for nrgrep_coords to search for matches.  
Usage: perl unjustify_fasta.pl <file1> <file2> <file3>...  
Example: perl unjustify_fasta.pl ATH1.pep ATH1.cds  
         creates in the same directory ATH1.pep.prepared and ATH1.cds.prepared

Run PatMatch with the command:  

perl patmatch.pl [residue] [pattern] [sequence file] [number of mismatches] [mismatch types]

Residue types:  

-n for nucleotide  
-c for complementary strand  
-p for protein  

Mismatch types can be:

ids for insertions, deletions, and substitutions  
i   for insertions only  
id  for insertions and deletions  
etc.  


Pattern Syntax
--------------

This program uses the same pattern syntax as the version of Patmatch available  
on TAIR's website, http://arabidopsis.org/cgi-bin/patmatch/nph-patmatch.pl  

The website contains information about the supported pattern syntax as well as  
some example patterns.  

When running a pattern, if you get an "Invalid pattern syntax.  Please check your pattern and try again." error,  
and you are certain of the correctness of your pattern, try enclosing just the pattern in double quotes, such as:  

perl patmatch/patmatch.pl -p "L{3,5}X{5}DGZ" patlibs/ATH1_pep 1 ids  

for an example search on a peptide pattern with 1 insertion/deletion/substitution allowed.

## About setting up this repository and using this interface

This particular notebook page runs with access directly to the Bash shell. You can type in commands in the cells as if you were on the command line.

If you'd like a more full-featured Jupyter environment, proceed onto [the next notebook page](PatMatch%20with%20Python%20basics.ipynb) where the cells are connected to the interactive Python shell, or press the `Jupyter` symbol in the upper, left at the top of this page and then click on `New` on the right above the directory listing, and selecting a Python notebook from the drop-down menu. In that interactive Python-based notebook, you'll need to either preface your code to be directed to the command line with exclamation points. Alternatively, entire cells can be designated be run as Bash script with the `%%bash` magic tag as the first line of the cell. Learn about that and many of the magics commands [here](https://www.dataquest.io/blog/jupyter-notebook-tips-tricks-shortcuts/).
In order to set up an environment with Perl and the required modules specified, the `environment.yml` file was edited. The software was copied from ftp://ftp.arabidopsis.org/home/tair/Software/Patmatch/ to the repository, and during the spinning up of the Binder instance, the process triggers via the `postBuild` file unpacking and compiling using `make` running in the directory generated in the course of the unpacking.
